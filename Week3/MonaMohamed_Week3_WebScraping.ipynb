**- Import the required lib**

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

**- Call the url and read the table**

```python
url = 'https://www.worldometers.info/coronavirus/country/us/#nav-yesterday'
resp = requests.get(url)
data = resp.text # parse HTTP response

soup = BeautifulSoup(data, 'html.parser')
table = soup.find(id="usa_table_countries_yesterday")

df = pd.read_html(str(table))[0]
```

**- Dataframe column names should match the HTML column headers**


```python
#print out DF info df
df.info()
```

**- Print out the DF**

```python
# read all data in DF
#display(df)
display(df)
```

```python
# read first 10 
df.head(10)
```

**- Eexclude the '#', 'source', and 'projections' columns**

```python
# remove #, and Projections columns
# no Resource column in the yesturday table
df.drop(columns=['#', 'Projections'], inplace= True)

# read first 10 
df.head(10)
```

**- Exclude rows for country totals**

```python
# remove USA total and Total rows
df.drop([0, 64], inplace = True)

# display df to make sure no total row left 
display(df)
```

```python
# State column is only names no HTML code
# read first 10 
df.head(10)
```

# 4.  Print the the top 5 states (name & value) for each of these metrics:

 **- New cases**

```python
# Sort df by NewCases values descending and print out the first maximum 5, show only USAState and NewCases columns 
df[['USAState', 'NewCases']].sort_values(by=['NewCases'], ascending= False).head(5)
```

 **- Total deaths**

```python
# Sort df by TotalDeaths values descending and print out the first maximum 5, show only USAState and TotalDeaths columns 
df[['USAState', 'TotalDeaths']].sort_values(by=['TotalDeaths'], ascending= False).head(5)
```

**- Total cases / 1M pop**

```python
# Show the columns list before renaming
list(df.columns)
```

```python
# Rename the 'Tot Cases/1M pop' column as it contain a non-breaking space 'Tot\xa0Cases/1M pop'
df.rename(columns={'Tot\xa0Cases/1M pop': 'Tot Cases/1M pop'}, inplace=True)

# Show list of the new columns name to make sure it has been renamed
list(df.columns)
```

```python
# Sort df by Tot Cases/1M pop values descending and print out the first maximum 5, show only USAState and Tot Cases/1M pop columns 
df[['USAState', 'Tot Cases/1M pop']].sort_values(by=['Tot Cases/1M pop'], ascending= False).head(5) 

```

**- Total deaths / 1M pop** 

```python
# Sort df by Deaths/1M pop values descending and print out the first maximum 5, show only USAState and Deaths/1M pop columns 
df[['USAState', 'Deaths/1M pop']].sort_values(by=['Deaths/1M pop'], ascending= False).head(5)
```
